```python
import pandas as pd
from pathlib import Path
import hvplot.pandas
```

```python
# Display July 2020 Music Royalty Payment Statement

SampleRoyaltyPayment = pd.read_csv(
    Path('./SampleRoyaltyPayment.csv'))

#SampleRoyaltyPayment = SampleRoyaltyPayment.set_index('Distribution Date')

display(SampleRoyaltyPayment.head())
display(SampleRoyaltyPayment.tail())
```

```python
# Summarize the Royalty Data by all Sources and Respective Music Users

SampleRoyaltyPayment_ByMusicUser = SampleRoyaltyPayment[["Source", "Music User","Total Payable"]].groupby(["Source","Music User"]).sum()
display(SampleRoyaltyPayment_ByMusicUser)
```

```python
# Summarize the Royalty Data by Unique Sources Only:

SampleRoyaltyPayment_BySource = SampleRoyaltyPayment[["Source", "Total Payable"]].groupby(["Source"]).sum()
display(SampleRoyaltyPayment_BySource)
```

```python
# Chart Royalty Data by Sources Only:

SampleRoyaltyPayment_BySource.hvplot.bar(
    xlabel = "Music User",
    ylabel = "Total Payable",
    title="Royalty Payment From Different Sources",
    rot=45)
```

```python
# Summarize Royalty Data by Users in Streaming:

SampleRoyltyPayment_StreamOnly = SampleRoyaltyPayment[SampleRoyaltyPayment['Source'] == 'STREAMING']

SampleRoyltyPayment_StreamOnly = SampleRoyltyPayment_StreamOnly[["Music User","Total Payable"]].groupby(["Music User"]).sum()
display(SampleRoyltyPayment_StreamOnly)
```

```python
# Chart Royalty Data by Music Users in Streaming:

SampleRoyltyPayment_StreamOnly.hvplot.bar(
    xlabel = "Music User",
    ylabel = "Total Payable",
    title="Royalty Payment From Different Streaming Channels",
    rot=90)
```
